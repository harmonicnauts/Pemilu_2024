## Fetch dan convert data dari API punya bang [@zakiego](https://twitter.com/zakiego) menjadi file csv

* url_api : [https://caleg.zakiego.com/api](https://caleg.zakiego.com/api)
* url_artikel : [https://zakiego.com/articles/pemilu-2024](https://zakiego.com/articles/pemilu-2024)

```python
import requests
```

```python
def fetch_data(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            print("Fetched data:")
            print(response.text)
            return response.text
        else:
            print(f"Failed to fetch data. Status code: {response.status_code}")
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
url_to_fetch = "https://caleg.zakiego.com/api/"
url_suffix='/dapil/list'
root = fetch_data(url_to_fetch)
```

```python
type(root)
```

```python
# conv data from str -> json format
import json

root_json = json.loads(root)
print(type(root_json))
```

```python
print(root_json['data'])
for i in root_json['data']:
  print(i)
```

```python
def fetch_data_json(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            print("Fetched data:")
            print(response.text)
            return json.loads(response.text)
        else:
            print(f"Gagal fetch data. Status code: {response.status_code}")
    except Exception as e:
        print(f"error : {e}")
```

```python
output_json = {} #object var

for kategori in root_json['data']:
  url = url_to_fetch + kategori + url_suffix
  temp = fetch_data_json(url)
  temp_data = temp['data']

json_output = json.dumps(output_json)
```

```python
import pandas as pd
output_json = {} #object var

df_caleg = pd.DataFrame({
    "id_dapil": [],
    "nama_dapil": [],
    "kategori": [],
    "url_dapil": []
})

for kategori in root_json['data']:
  url = url_to_fetch + kategori + url_suffix
  temp = fetch_data_json(url)
  temp_data = temp['data']

  print(type(temp_data))
  print(temp_data[0]['namaDapil'])

  for element in temp_data:
    print(element)
    id_dapil = element['id']
    nama_dapil = element['namaDapil']
    url_dapil = element['url']

    new_row = pd.DataFrame({
      'id_dapil':[id_dapil],
      'nama_dapil' : [nama_dapil],
      'kategori' : [kategori],
      'url_dapil' : [url_dapil],
    })
    # print(new_row)
    df_caleg = pd.concat([df_caleg, new_row], ignore_index=True)
```
